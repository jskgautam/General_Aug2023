##1. Using locals() and globals() library methods.

```python
var_global1 = 5
def funcD():
  var_local1 = 1
  # Check Local variable.
  if 'var_local1' in locals():
    print('The variable,' + ' var_local1, '+ 'is defined and it is local!')
  else:
    print('The variable,' + ' var_local1, '+ 'is NOT defined!')

  if 'var_local2' in locals():
    print('The variable,' + ' var_local2, '+ 'is defined!')
  else:
    print('The variable,' + ' var_local2, '+ 'is NOT defined!')

  # Check Global variable.
  if 'var_global1' in globals():
    print('The variable,' + ' var_global1, '+ 'is defined and it is global!')
  else:
    print('The variable,' + ' var_global1, '+ 'is NOT defined!')

  if 'var_globalX' in globals():
    print('The variable,' + ' var_globalX, '+ 'is available and it is defined!')
  else:
    print('The variable,' + ' var_globalX, '+ 'is NOT defined!')
funcD()
# For a global variable you can also test it outside of the function.
```

##2. Using Try-except method

```python
var_global3 = 4
def funcE():
  var_local3 = 6

  # Check Local variable.
  try:
    var_local3
    print('The variable,' + ' var_local3, '+ 'is defined and it is local!')
  except NameError:
    print('The variable,' + ' var_local3, '+ 'is NOT defined!')

  try:
    var_local4
    print('The variable,' + ' var_local4, '+ 'is defined and it is local!')
  except NameError:
    print('The variable,' + ' var_local4, '+ 'is NOT defined!')

  # Check Global variable.
  try:
    var_global3
    print('The variable,' + ' var_global3, '+ 'is defined and it is global!')
  except NameError:
    print('The variable,' + ' var_global3, '+ 'is NOT defined!')
  try:
    var_global4
    print('The variable,' + ' var_global4, '+ 'is defined and it is global!')
  except NameError:
    print('The variable,' + ' var_global4, '+ 'is NOT defined!')
funcE()
# For a global variable you can also test it outside of the function.
```
